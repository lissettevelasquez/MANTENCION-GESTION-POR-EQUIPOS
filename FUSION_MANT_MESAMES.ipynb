```python
import pandas as pd

# Cargar el archivo de Excel
file_path = "/content/MANT-09.XLSX"
xls = pd.ExcelFile(file_path)
```

```python

# Cargar las dos hojas
df_ot = xls.parse('OT')
df_gasto = xls.parse('GASTOS')
```

```python
# Asegurar que la columna "Orden" no tenga espacios
df_ot.columns = df_ot.columns.str.strip()
df_gasto.columns = df_gasto.columns.str.strip()
```

```python

# Fusionar las tablas por "Orden" (outer join para conservar todos los datos)
df_merged = pd.merge(df_ot, df_gasto, on="Orden", how="outer", suffixes=("_OT", "_GASTOS"))
```

```python

# Guardar el archivo fusionado
output_path = "/content/datos_MANT09.xlsx"
df_merged.to_excel(output_path, index=False)
```

```python
# Permitir descarga del archivo fusionado
from google.colab import files
files.download(output_path)
```
